# MNIST Exploratory Data Analysis

```python
# Step 1: Import required libraries
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 2: Load and normalize the MNIST dataset
# This gives us 60,000 training images and labels
(x_train, y_train), _ = tf.keras.datasets.mnist.load_data()

# Step 3: Plot the first 9 digit samples to get a visual feel for the dataset
plt.figure(figsize=(6, 6))  # Set the overall figure size
for i in range(9):
    plt.subplot(3, 3, i + 1)  # Arrange plots in a 3x3 grid
    plt.imshow(x_train[i], cmap='gray')  # Show the image in grayscale
    plt.title(f"Label: {y_train[i]}")  # Show the digit label as title
    plt.axis('off')  # Hide the axis ticks
plt.tight_layout()
plt.show()
```

```python

# Step 4: Plot the distribution of digit labels
# This tells us how many of each digit we have in the training data
plt.figure(figsize=(8, 4))
plt.hist(y_train, bins=np.arange(11)-0.5, edgecolor='black', rwidth=0.8)
plt.xticks(range(10))  # Show digit labels 0–9
plt.xlabel("Digit Label")
plt.ylabel("Frequency")
plt.title("Label Distribution in MNIST Training Set")
plt.grid(True)
plt.show()
```

```python
# This shows how many pixels fall into different brightness ranges (0 = black, 1 = white)
plt.hist(x_train[0].flatten(), bins=25, color='purple')
plt.title("Pixel Intensity Distribution for First Digit")
plt.xlabel("Pixel Value")
plt.ylabel("Count")
plt.show()
```

```python
# This creates an 'average' image for each digit from 0 to 9
# We take all images of '0', all images of '1', etc., and compute their mean pixel-wise
avg_digits = np.zeros((10, 28, 28))  # Container for 10 averaged images

for digit in range(10):
    avg_digits[digit] = x_train[y_train == digit].mean(axis=0)

# Plot the average image for each digit
plt.figure(figsize=(12, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(avg_digits[i], cmap='gray')
    plt.title(f"Avg: {i}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
