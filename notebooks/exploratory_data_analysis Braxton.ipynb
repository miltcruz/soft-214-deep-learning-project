# üß† Project Goal: MNIST Digit Classifier ‚Äì Exploratory Data Analysis (EDA)

**MNIST (Modified National Institute of Standards and Technology)** is a classic benchmark dataset in machine learning.  
It contains:
- üñºÔ∏è 60,000 training images and 10,000 test images
- üßÆ Grayscale handwritten digits (0 through 9)
- üìè Each image is 28√ó28 pixels

The goal is to train a model that can **accurately recognize digits** from these images.

## üîç What is EDA (Exploratory Data Analysis)?

**Exploratory Data Analysis (EDA)** is the process of:
- Summarizing the main characteristics of a dataset
- Visualizing patterns, trends, and anomalies
- Understanding structure and quality before modeling

## üéØ Project Objective

To explore the MNIST dataset and gain insights before training a machine learning model.  
This includes:

- Visualizing sample digit images (0‚Äì9)
- Checking class balance across digit labels
- Understanding pixel intensity patterns
- Identifying any anomalies or quality issues

üìå This is **Phase 1** of building a complete digit recognition pipeline using Python, NumPy, Matplotlib, Seaborn, and machine learning tools.


```python
# Libraries
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
```

```python
# Load the MNIST dataset
(x_train, y_train), _ = tf.keras.datasets.mnist.load_data() 
```

```python
# Display the first 9 images in a 3x3 grid
plt.figure(figsize=(6,6))
for i in range(9):
    plt.subplot(3,3,i+1) #Display 9 images  in a 3x3 grid
    plt.imshow(x_train[i], cmap='gray') # Display the image in grayscale
    plt.title(f'Label: {y_train[i]}')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Plot label distribution histogram using Matplotlib
plt.figure(figsize=(8, 4))
plt.hist(y_train, bins=np.arange(11)-0.5, edgecolor='black', rwidth=0.8)
plt.xticks(range(10))
plt.title('Distribution of Digit Labels in MNIST Training Set')
plt.xlabel('Digit Labels')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Visual Pie Chart of Label Distribution
# Get label counts
labels, counts = np.unique(y_train, return_counts=True)

# Define colors and explode effect for fun
colors = plt.cm.tab10.colors
explode = [0.05] * 10  # Slightly separate each slice

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    counts,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode=explode,
    shadow=True
)
plt.title("üìä MNIST Digit Distribution ‚Äì Pie Chart")
plt.axis('equal')  # Equal aspect ratio makes the pie circular
plt.show()
```
